```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
# Define the file path
file_path = "/content/drive/MyDrive/EDA/revenuefinal.csv"

# Import the dataset
df = pd.read_csv(file_path)

# Display the first few rows to verify import
df.head()
```

```python
# Display the column names
print(df.columns)
```

```python
# Display the shape of the dataset
print("Number of rows:", df.shape[0])
print("Number of columns:", df.shape[1])
```

```python
# Display the dataset information
print(df.info())
```

```python
# Display unique categories
unique_categories = df['Categories'].unique()
print("Unique categories:", unique_categories)
```

```python
# Summary statistics for numerical columns
print(df.describe())
```

```python
# Check for missing values
missing_values = df.isnull().sum()
print("Missing values per column:")
print(missing_values)
```

```python
# Filter companies from the finance category
finance_companies = df[df['Categories'] == 'Finance']
print(finance_companies.head())
```

```python
# Sort the dataset by Rank in ascending order
sorted_df = df.sort_values(by='Rank', ascending=True)
print(sorted_df.head())
```

```python
# Calculate total revenue
total_revenue = df['Revenue'].sum()
print("Total revenue generated:", total_revenue)
```

```python
# Sort the dataset by Marketcap in descending order and select the top 10
top_10_marketcap = df.sort_values(by='Marketcap', ascending=False).head(10)

# Plot the bar chart
plt.figure(figsize=(10, 6))
plt.bar(top_10_marketcap['Names'], top_10_marketcap['Marketcap'], color='skyblue')
plt.title('Top 10 Companies with Highest Market Capitalization')
plt.xlabel('Company Names')
plt.ylabel('Market Capitalization')
plt.xticks(rotation=90)
plt.show()
```

```python
# Plot the distribution of companies based on revenue
plt.figure(figsize=(20, 6))
plt.hist(df['Revenue'], bins=20, color='orange', edgecolor='black')
plt.title('Distribution of Companies Based on Revenue')
plt.xlabel('Revenue')
plt.xticks(rotation=90)
plt.ylabel('Number of Companies')
plt.show()
```

```python
# Plot the share prices of different companies
plt.figure(figsize=(15, 20))
plt.bar(df['Names'], df['Share.Price'], color='green')
plt.title('Share Prices of Different Companies')
plt.xlabel('Company Names')
plt.ylabel('Share Price')
plt.xticks(rotation=90)
plt.show()
```

```python
# Plot the distribution of companies based on ranks
plt.figure(figsize=(20, 6))
plt.hist(df['Rank'], bins=20, color='purple', edgecolor='black')
plt.title('Distribution of Companies Based on Rank')
plt.xlabel('Rank')
plt.xticks(rotation=90)
plt.ylabel('Number of Companies')
plt.show()
```

```python
# Sort the dataset by Revenue in descending order and select the top 10
top_10_revenue = df.sort_values(by='Revenue', ascending=False).head(10)

# Plot the horizontal bar chart
plt.figure(figsize=(10, 6))
plt.barh(top_10_revenue['Names'], top_10_revenue['Revenue'], color='red')
plt.title('Top 10 Companies with Highest Revenue')
plt.xlabel('Revenue')
plt.ylabel('Company Names')
plt.show()
```

```python
# Count the number of companies from each country
country_counts = df['Country'].value_counts()

# Plot the bar chart
plt.figure(figsize=(10, 6))
country_counts.plot(kind='bar', color='green')
plt.title('Distribution of Companies by Country')
plt.xlabel('Countries')
plt.ylabel('Number of Companies')
plt.xticks(rotation=45)
plt.show()
```

```python
# Extract numerical values from 'Revenue' column
df['Revenue'] = df['Revenue'].str.replace(r'\D', '', regex=True).astype(float)

# Define revenue ranges (you can adjust these as needed)
low_revenue_range = (0, 1000)
medium_revenue_range = (1001, 10000)
high_revenue_range = (10001, df['Revenue'].max())

# Categorize companies based on revenue ranges
df['RevenueRange'] = pd.cut(df['Revenue'], bins=[low_revenue_range[0], medium_revenue_range[0], high_revenue_range[0], high_revenue_range[1]], labels=['Low', 'Medium', 'High'])

# Count the number of companies in each revenue range
revenue_range_counts = df['RevenueRange'].value_counts()

# Create a pie chart
fig = px.pie(revenue_range_counts, values=revenue_range_counts.values, names=revenue_range_counts.index, title='Distribution of Companies by Revenue Range')
fig.show()
```

```python
# Select the top 5 categories by count
top_categories = df['Categories'].value_counts().nlargest(5)

# Create a pie chart
fig = px.pie(top_categories, values=top_categories.values, names=top_categories.index, title='Top 5 Categories of Companies')
fig.show()
```

```python
# Exclude non-numeric columns from the correlation calculation
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Plot heatmap for the numeric columns
plt.figure(figsize=(12, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap for the Numeric Columns')
plt.show()
```

```python
# Create a pairplot for the entire dataset
sns.pairplot(df)
plt.suptitle('Pairplot for the Entire Dataset', y=1.02)
plt.show()
```
